# Steps to build a Model in Python
## 1) Import your libraries
## 2) Imprt your data
## 3) Train a model (today we train a Tree)
## 4) Test your model (Visualize it)

# 1) Import required libraries

```python
# Machine Learning Packages
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO 
#Visualization Packages
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

# Iris Flowers

```python
Image(filename='/Users/ghorbaj/Google Drive/GA/D3 - Decision Tree PPT + Jypyter/iris.jpg')
```

# 2) Import the data

```python
iris=datasets.load_iris()
print('flower names are: ',iris.target_names)
print('\nThe data avilable for each flower: ','\n - sepal length in cm        \n - sepal width in cm        \n - petal length in cm       \n - petal width in cm')
```

Lets see how the data looks like:

```python
flower_data=pd.DataFrame(iris.data, columns=iris.feature_names)
flower_names=iris.target 
print('5 example flowers:  \n (X0, X1, X2, X3) \n',iris.data[47:52])
print('\nType of these 5 flowers: \n',iris.target[47:52])
# ['setosa' -->0 'versicolor'-->1 'virginica'-->2]
```

# 3) Train a tree classifier 

```python
tree=DecisionTreeClassifier()
tree.fit(flower_data,flower_names)
```

# 4) Visualize the tree

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data,filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
# ['setosa' --> Orange 
#  'versicolor'-->Green 
#  'virginica'-->Purple ]
```

# Predict a flower type

```python
#           [X0 ,  X1, X2,   X3 ]
test_flower=[0.1, 0.5,  1,  1]

tree.predict([test_flower])
# ['setosa' -->0 'versicolor'-->1 'virginica'-->2]
```
